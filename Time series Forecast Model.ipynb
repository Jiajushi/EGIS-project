## Using additive model to predict

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

daily = df_daily.resample(rule='D').sum()

result = seasonal_decompose(daily, model = 'additive') 
plt.rcParams['figure.figsize'] = (20, 9)
result.plot();
```

### Stationarity test

```python
from statsmodels.tsa.stattools import adfuller

adfuller(daily)

#P value<5%, stable
```

### Acf & Pacf test

```python
plt.rc("figure", figsize=(18,4))
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(daily); 
plot_pacf(daily);
```

## Forecast & Prophet model

```python
import pandas as pd
from prophet import Prophet

daily_prophet_df = pd.DataFrame(daily) 
daily_prophet_df.reset_index(inplace=True)
daily_prophet_df.columns = ['ds', 'y'] 
daily_prophet_df
```

```python
train_samples = 360

train_set = daily_prophet_df.iloc[:-360] 
test_set = daily_prophet_df.iloc[-360:]

print("Training set: ", train_set.shape[0])
print("Test set: ", test_set.shape[0])
```

```python
model = Prophet()
model.fit(train_set)
```

```python
future = model.make_future_dataframe(periods= 360,freq = 'd') 
future.tail()
```

```python
forecast = model.predict(future)
forecast.columns
```

```python
preds = pd.DataFrame({'Prediction': forecast.yhat[-360:]}) 
preds.index = pd.to_datetime(forecast.ds[-360:]) 
preds.index.names = ['Date'] 
preds
```

```python
test_set['y']
```

```python
preds['Prediction']
```

```python
y_test = test_set.set_index('ds')['y'] 
y_pred = preds['Prediction'] 
pd.DataFrame({'y_test': y_test, 'y_pred' : y_pred, 'diff':y_test - y_pred})
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

print('MSE = ', mean_squared_error(y_test,y_pred)) 
print('RMSE = ', np.sqrt(mean_squared_error(y_test,y_pred)))
print('MAE = ', mean_absolute_error(y_test,y_pred))
```
